# Queno 2

# Objective Function of Linear SVM

```python
#The objective of linear SVM is to find a hyperplane that maximizes the margin that is distance between hyperplane and 
#closest data point between two classes under the constraint that all datapoints lie on correct side of hyperplane.
```

```python
#Hard margin SVM imposes that all datapoints must be outside the area between margin lines.
```

```python
#So objective is to maximize the margin under the constraints that all datapoints lie on the correct side of hyperplane.
```

# Queno 3

# Kernel trick in SVM

```python
#Kernel trick is a method used in Support Vector Machines to convert data into a higher dimensional feature space where 
#it may be linearly separated.
```

```python
#This technique enables SVM to identify a hyperplane that separates the data with maximum margin even when data is not 
#linearly separable in its original space. 
```

```python
#For example lets take two class red and blue which is in 2d space and not linearly separable.
```

```python
#To make this data linearly separable we can use kernel trick.
```

```python
#It enables SVM to solve non linear classification problems by mapping input data to high dimensional feature space.
```

```python
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
for keys in iris.keys() :
    print(keys)
```

```python
X=iris.data
y=iris.target
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5) 
#So we are splitting haif of data into training set and half into test set.
```

```python
X_train.shape
X_test.shape
```

```python
y_train.shape
y_test.shape
```

```python
print('X_train')
print(X_train)
print('X_test')
print(X_test)
```

```python
from sklearn import svm
```

```python
clf=svm.SVC(kernel='linear',C=1.0)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
y_pred
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
#So we get classification rate of 97.33%
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
w=clf.coef_[0]
a=-w[0]/w[1]
XX=np.linspace(0,13)
yy=a*XX-clf.intercept_[0]/w[1]
plt.plot(XX,yy,'k-')
plt.scatter(X_train[:,0],X_train[:,1],c=y_train)
plt.legend()
plt.show()
```

```python
clf=svm.SVC(kernel='linear',C=5.0)
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test) # Changes we see after put changing parameter 'C'.
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
w=clf.coef_[0]
a=-w[0]/w[1]
XX=np.linspace(0,13)
yy=a*XX-clf.intercept_[0]/w[1]
plt.plot(XX,yy,'k-')
plt.scatter(X_train[:,0],X_train[:,1],c=y_train)
plt.legend()
plt.show()
```
