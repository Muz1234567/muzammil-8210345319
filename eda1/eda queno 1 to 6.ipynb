# Queno 6

```python
#First of all we have to normalize dataset of wine quality.
```

```python
normalized_df
```

```python
#now we can convert in to array
normalized=normalized_df.to_numpy()
```

```python
#Now we can perform PCA on this normalized data.
from sklearn.decomposition import PCA
pca=PCA()
proj_wine=pca.fit_transform(normalized)
```

```python
proj_wine.shape
```

```python
import numpy as np
var=np.sum(pca.explained_variance_[0:8])
#on selecting principal component 8 we are able to get amount of variance below
```

```python
var
```

```python
var_percentage=np.sum(pca.explained_variance_ratio_[0:8])*100
```

```python
var_percentage
#on selecting 8 principal component we are able to analyze 90% of variance.
```

```python
#so minimum pricipal component required to explain 90% of variance of dataset is 8.
```
