```python
import matplotlib.pyplot as plt
```

```python
rating=googleplaystore['Rating'].head(35)
size=googleplaystore['Size'].head(35)
```

```python
plt.figure(figsize=(15,10))
plt.scatter(size,rating)
plt.xlabel('size')
plt.ylabel('rating')
plt.title('Effect of app size on rating')
```

So in this scatter plot  of 40 app we see that different size of app have different rating as we see app with 2.7m size
have very low rating that is 3.2 and app with higest rating 4.8 have size between 6m to 6.1m while app with largest size
that 39m have rating only 4 which shows that app sizedoes not have significant effect on app rating.

# Queno 16

We can use rating,reviews and  installs feature to know most succesful app developers.

```python
googleplaystore['Reviews']=googleplaystore['Reviews'].str.replace(',','').astype(int)
```

```python
reviews=googleplaystore['Reviews']
installs=googleplaystore['Installs']
rating=googleplaystore['Rating']
category=googleplaystore['Category']
```

```python
plt.figure(figsize=(20,15))
plt.bar(category,installs)
plt.xlabel('category')
plt.ylabel('no of install')
plt.title('Distribution of app installation number with respect to category')
```

So with respect to category communication,entertainment and books and reference category app have high instalation number.

```python
plt.figure(figsize=(20,15))
plt.bar(category,reviews)
plt.xlabel('category')
plt.ylabel('Rviews')
plt.title('Distribution of app reviews number with respect to category')
```

So communication category app has higher reviews.

```python
plt.figure(figsize=(20,15))
plt.bar(category,rating)
plt.xlabel('category')
plt.ylabel('Rating')
plt.title('Distribution of app rating number with respect to category')
```

According to rating comics and dating app category have higher rating.

# Analysis

From the above observation we can conclude that app developer with communication category is the most succesful
app developer as they are higher in installation number and higher reviews than other app category.

# Queno 17

We can use last updated feature to analyse best time to launch new app.

```python
frequency_tab=pd.crosstab(index=googleplaystore['Installs'],columns=googleplaystore['Last Updated'])
```

```python
frequency_tab
```

```python
result=frequency_tab.sort_values(by='Installs',ascending=False)
```

```python
pd.set_option('display.max_columns',100)
result
```

```python
So according to frequency table August month got higher installation in year 2016,2017,2018 and so on so August month is 
best time to launch new app.
```
