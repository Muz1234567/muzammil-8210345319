```python
#Column price may have outliers as its mean value that is 46,787 and max value 95,657 so we find outliers with help 
#of defined function
```

```python
def find_outliers_IQR(flight_price):
    q1=flight_price.quantile(0.25)
    q3=flight_price.quantile(0.75)
    IQR=q3-q1
    outliers=flight_price[((flight_price<(q1-1.5*IQR))|(flight_price>(q3+1.5*IQR)))]
    return outliers
```

```python
outliers=find_outliers_IQR(flight_price['price'])
print("number of outliers:"+str(len(outliers)))
print("max outlier value:"+str(outliers.max()))
print("min outlier value:"+str(outliers.min()))
```

```python
import numpy as np
q1,q3=np.percentile(flight_price['price'],[25,75])
iqr=q3-q1
lower_bound=q1-1.5*iqr
upper_bound=q3+1.5*iqr
outliers=flight_price['price'][(flight_price['price']<lower_bound)|(flight_price['price']>upper_bound)]
```

```python
outliers
```

So with the above methods we are able to find outliers that is 29 in which 95,657 is largest outlier and 77,737 is
minimum outliers.So we conclude that outliers exist in higher range not in lower range of dataset and we can see it 
using boxplot.

```python
flight_price.boxplot(column=['price'])
```

So in this figure we wee all outliers in form of bubble exist between 77737 to 95,657.

There are various variables that can affect flight price dataset are from,to ,total time taken,stop arethe columns that 
can affect the flight price.

```python
flight_price['stop']
```

```python
flight_price['no of stop']=flight_price['stop'].astype('str').str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)
```

```python
flight_price
```

```python
flight_price=flight_price.fillna(0)
```

```python
flight_price
```

```python
sns.catplot(x='no of stop',y='price',data=flight_price)
```

So here we have plotted category plot with respect to no of stop as we see that stop with 1 have higher and lower
 flight prices while stop with 0 only have lower flight price as compared to flight having 1 stop.
