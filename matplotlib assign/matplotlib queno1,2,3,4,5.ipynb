# Scatter Plot

```python
#Scatter plots are the graphs that present the relationship between two variables in a data-set.
```

```python
#It represents data points on two dimensional plane, the independent variable or attribute is plotted on X-axis,while the dependent variable is plotted on Y-axis.
```

```python
#It is used when we have paired numerical data or when there are multiple values of the dependent variable for a unique value of an independent variable.
```

```python
import numpy as np
```

```python
np.random.seed(3)
```

```python
x=3+np.random.normal(0,2,50)
```

```python
y=3+np.random.normal(0,2,len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,3))
plt.scatter(x,y,c='g',marker='v')
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.title("this is my scatter plot X vs Y")
plt.grid()
```

# Sub Plot Function

```python
#Sub plot function is used to show multiple line plots in same figure.
```

```python
#plot1

plt.figure(figsize=(10,5))
x=np.array([0,1,2,3,4,5])
y=np.array([0,100,200,300,400,500])
plt.subplot(2,2,1)
plt.title("this is plot1")
plt.grid()
plt.plot(x,y)
#plot2
x=np.array([0,1,2,3,4,5])
y=np.array([50,20,40,20,60,70])
plt.subplot(2,2,2)
plt.title("this is plot2")
plt.grid()
plt.plot(x,y)
#plot3
x=np.array([0,1,2,3,4,5])
y=np.array([10,20,30,40,50,60])
plt.subplot(2,2,3)
plt.title("this is plot3")
plt.grid()
plt.plot(x,y)
#plot4
x=np.array([0,1,2,3,4,5])
y=np.array([200,350,250,550,450,150])
plt.subplot(2,2,4)
plt.title("this is plot4")
plt.grid()
plt.plot(x,y)
plt.show()
```

# Bar Plot

```python
#A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights proportional to values which they represent. 
```

```python
#It can be plotted horizontally or vertically.
```

```python
#It is used to visualize the categorical data.
```

```python
company=np.array(["Apple","Microsoft","Google","AMD"])
profit=np.array([3000,8000,1000,10000])
```

```python
company
```

```python
profit
```

```python
plt.figure(figsize=(6,3))
plt.bar(company,profit)
plt.xlabel("Comapany name")
plt.ylabel("Profit")
plt.title("Company and its profit")
plt.grid()
```

```python
plt.barh(company,profit)
plt.title("Company and its profit in horizontal bar")
plt.xlabel("Profit")
plt.ylabel("Company name")
plt.grid()
```

# Box Plot

```python
#A box plot is a way to visualise the distribution of data by using box and some vertical lines. 
```

```python
#The data can be distributed between five key ranges,which are as follows:
#1.Minimum
#2.1st quartile
#3.Median
#4.3rd quartile
#5.Maximum
```

```python
#It is used to display a summary of dataset having above properties.
```

```python
box1=np.random.normal(100,10,200)
box2=np.random.normal(90,20,200)
```

```python
box1
```

```python
box2
```

```python
np.random.seed(10)
box1=np.random.normal(100,10,200)
box2=np.random.normal(90,20,200)
box=[box1,box2]
fig=plt.figure(figsize=(10,5))
plt.boxplot(box)
plt.xlabel("No of boxes")
plt.ylabel("Boxes range")
plt.title("Plotting box")
plt.grid()
plt.show()
```
