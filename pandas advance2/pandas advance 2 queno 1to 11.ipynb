# queno 6

```python
#Windows functions are used to perform some sort of calculations across a set of rows that are interrelated with current row.
```

```python
#Windows function in pandas returns a value for every row of the original table.
```

```python
#Window is a value(in terms of row) that tells us the distance between our current row and the row which is used for calculation.
```

```python
#There are many functions can be performed using windows.
```

```python
data=[10,20,30,40,50,60]
df2=pd.DataFrame(data,columns=['Numbers'])
```

```python
df2
```

```python
#now use windows rolling function in it.
```

```python
df2.Numbers.rolling(1).sum()
#here we have size 1 so it is adding by taking data one by one.
```

```python
df2.Numbers.rolling(3).sum()
#in this we have first two row as nan because it is taking at least 3 data to add so in row 1 and 2 data is not enough so it gives nan.
```

```python
df2.Numbers.rolling(1).mean()
#calculating mean
```

```python
df2.Numbers.rolling(3).mean()
```

```python
df2.Numbers.rolling(4).min()
#Here it is inspecting four data and taking small data from them.
```

```python
df2.cumsum()
#Here it is having cumulative sum means adding whole number with current value.
```

```python
#so in this way we are able to do different kinds of function with window function.
```
