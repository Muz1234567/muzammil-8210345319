```python
from flask import Flask,request
```

```python
app=Flask(__name__)
@app.route('/language_type')
def language_type():
    language=request.args.get('language')
    return '<h1>The language is: {}</h1>'.format(language)
if __name__=='__main__':
    app.run()
```
