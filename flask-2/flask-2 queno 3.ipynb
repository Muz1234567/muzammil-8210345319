```python
from flask import Flask,redirect,url_for
app=Flask(__name__)
```

```python
@app.route("/")
@app.route("/home")
def home():
    return "this is my home page"
    
```

```python
@app.route("/login")
def login():
    return "login successfully.."
```

```python
@app.route("/logout")
def logout():
    return redirect(url_for("home"))
```

```python
if __name__=='__main__':
    app.run()
```

```python
#so in this way redirect() functions allows us to redirect the user to another url.
```

```python
#as in the above url logout redirected to home url.
```
