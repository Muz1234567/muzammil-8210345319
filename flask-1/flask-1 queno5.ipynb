```python
from flask import Flask,redirect,url_for
```

```python
app=Flask(__name__)
```

```python
@app.route('/name')
def student():
    return 'muzammil'
```

```python
@app.route('/failed')
def failed_student():
    return 'student %s as failed' % student
```

```python
@app.route('/user/<name>')
def student_result(name):
    if name =='name':
        return redirect(url_for('student'))
    else:
        return redirect(url_for('failed_student',student=name))
    
```

```python
if __name__=='__main__':
    app.run()
```
